## Word2Vect

```python
import gensim.models.keyedvectors as word2vec
import nltk
import re
from nltk.tag import pos_tag
from nltk.corpus import stopwords
stop_words = set()
for word in set(stopwords.words('english')):
  stop_words.add(word)
  stop_words.add(word.capitalize())

print (stop_words)
```

```python
model = word2vec.KeyedVectors.load_word2vec_format("GoogleNews-vectors-negative300.bin", binary=True)
```

```python
len(model.word_vec('office'))
```

```python
print(model.word_vec('office'))
```

```python
### PRINT ALL WORDS
user_post = df.answer[0].split()

input_word = []
for line in user_post: 
  line = re.sub(r'[^a-zA-Z ]', '', line)
  n_az = len(re.findall('[A-Za-z]', line))
  n_ko = len(re.findall('[ก-๙]', line))
  if n_az >= n_ko :
    filtered_sentence = [w for w in line.split() if not w in stop_words]
    tagged_sent = pos_tag(filtered_sentence)
    input_word = input_word + [word for word,pos in tagged_sent] #if pos == 'NNP'

input_word_list = pd.Series(input_word)


```

```python
len (input_word_list)
```

```python
### PRINT ALL VECTORS
input_word_list = pd.Series(input_word)
for word in input_word_list.unique():
  string_d = ""
  #print (word)
  if word in model.wv.vocab:
    for vector in model.word_vec(word):
      string_d = string_d + str(vector) + "\t"
      
    print (string_d)
```

```python
input_word_list = pd.Series(input_word)
final_vector_array = []
final_word_array = []
for word in input_word_list.unique():
  #print (word)
  if word in model.wv.vocab:
    final_vector_array.append(model.word_vec(word))
    final_word_array.append(word)



```

```python
from sklearn.cluster import KMeans 
classifier_model = KMeans(n_clusters=4) 
classifier_model.fit(final_vector_array) 
```

```python
final_labels = classifier_model.predict(final_vector_array)
print(final_labels) 
```
